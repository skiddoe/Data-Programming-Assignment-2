### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import csv
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
url="https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users=pd.read_csv(url,delimiter='|')
users.head(5)
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation')[['age']].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
male=users[users.gender=='M'].occupation.value_counts()
female=users[users.gender=='F'].occupation.value_counts()
male_ratio=pd.DataFrame({'Male':male,'Female':female})
male_ratio['Male ratio']=male_ratio['Male']/(male_ratio['Male']+male_ratio['Female'])
male_ratio.drop('doctor', axis='rows', inplace=True)
male_ratio=male_ratio.sort_values('Male ratio', ascending=False)
male_ratio
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
age_max=users.groupby('occupation')[['age']].max()
age_min=users.groupby('occupation')[['age']].min()
age_maxmin=pd.concat([age_max,age_min], axis=1)
age_maxmin.columns=['Max age','Min age']
age_maxmin
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
multi_index=users.groupby(['occupation','gender']).age.mean()
multi_index=pd.DataFrame(multi_index)
multi_index.columns=['Mean age']
multi_index
```

### Step 8.  For each occupation present the percentage of women and men

```python
male=users[users.gender=='M'].occupation.value_counts()
female=users[users.gender=='F'].occupation.value_counts()
male_ratio=pd.DataFrame({'Male':male,'Female':female})
male_ratio['Male ratio']=male_ratio['Male']/(male_ratio['Male']+male_ratio['Female'])
female_ratio=pd.DataFrame({'Male':male,'Female':female})
male_ratio['Female ratio']=female_ratio['Female']/(male_ratio['Male']+male_ratio['Female'])
male_ratio.drop('doctor', axis='rows', inplace=True)
male_ratio=male_ratio.sort_values('Male ratio', ascending=False)
male_ratio
```
